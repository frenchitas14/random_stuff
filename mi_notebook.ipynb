```python
import pandas as pd 
import os 
import numpy as np 
import matplotlib as plt

cwd = os.getcwd()
print(cwd)

df = pd.read_csv("rdm_results.csv")
```

```python
n = len(df)
trials_per_coherence = df.groupby("coherence")["trial"].count()
acc = df["correct"].mean()*100
rt = df["decision_time_ms"].mean() 
accuracy_per_coherence = df.groupby("coherence")["correct"].mean()*100
rt_per_coherence = df.groupby("coherence")["decision_time_ms"].mean()

print(f"Number of trials = {n}")
print(f"Here is the number of trials per coherence = {trials_per_coherence}")
print(f"The overall accuracy is = {acc}")
print(f"The overall reaction time is = {rt:1f}")
print(f"Here is the mean accuracy per coherence level {accuracy_per_coherence}")
print(f"Here is the mean reaction time per coherence level {rt_per_coherence}")

table = pd.DataFrame({
    "Mean Accuracy (%)": accuracy_per_coherence,
    "Mean RT (ms)": rt_per_coherence
}).reset_index()

display(table)
```
